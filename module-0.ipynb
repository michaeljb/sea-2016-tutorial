# Python, Jupyter & pandas: Module 0

## Preparing the software

### Using the pre-built virtual machine (_prefered_)

This is the prefered method, but installing VirtualBox requires administrative priveleges on your system. If you cannot install VirtualBox, see if your system administrator can assist. If that fails, see the _Using Anaconda_ section below.

#### 1. Download and install VirtualBox

- Visit the [VirtualBox downloads page](https://www.virtualbox.org/wiki/Downloads) and get the latest VirtualBox platform package and Extension Pack.

- Install the VirtualBox platform package:
 
    - Linux: Get the `.deb` or `.rpm` file appropriate to your OS and install via your package manager (apt, yum, etc.).
 
    - OS X: Get the `.dmg` image, double click, and follow the prompts.
 
    - Windows: Get the `.exe` installer, double click, and follow the prompts.
 
    - See the [installation documentation](https://www.virtualbox.org/manual/ch02.html) if you have problems.

- Install the Extension Pack.

    - Launch VirtualBox, then click _File_ (Linux, Windows) or _VirtualBox_ (OS X), then _Preferences_ > _Extensions_. Click the _Adds new package_ button on the right, then navigate to and select the Extensions Pack file you downloaded earlier. Answer all questions in the affirmative to install the package.

#### 2. Get and run the virtual machine

_**Note:** The virtual-machine image file is ~1.35 GB in size and will take some time to download._

- Enter the following URL in your browser to download the virtual machine archive:

 `ftp://ftp.nsidc.org/pub/requests/pmadden/sea-2016-tutorial.zip`


- Extract the archive, open the extracted folder, and double-click `sea-tutorial-2016.vbox` to launch VirtualBox.

- Inside the VirtualBox manager, double-click the _sea-2016-tutorial_ machine to run it.

_When you're finished using the virtual machine, you can right-click on it in VirtualBox and select Close > Power Off._

#### 3. Connect to Jupyter

- When you see the virtual machine's `jupyter login:` prompt, navigate to [http://localhost:8888](http://localhost:8888) in your browser.

- Click the green _Start My Server_ button if you see it, and/or log in with username `jupyter` and password `jupyter`.

- Click to open the _sea-2016-tutorial_ link and you should see a list of the `module-*.ipynb` notebook files for the tutorial.

#### Updating the tutorial files

Update the tutorial files to the latest versions with the following steps (note that this will erase any changes you have made in the tutorial notebooks):

1. Click window created by VirtualBox when you started the virtual machine and, if you do not already see the login prompt, hit any key to wake the display.
2. Log in with username `jupyter` and password `jupyter`.
3. `cd ~/sea-2016-tutorial`
4. `git checkout . && git pull`
5. `logout`

### Using Anaconda (Linux and OS X only)

If you are using Linux or OS X, you may be able to use the Jupyter support in Anaconda.

#### 1. Download and install Anaconda

- [Download](https://www.continuum.io/downloads) the Anaconda (Python 3.5 version) package appropriate to your system and follow the [installation instructions](https://www.continuum.io/downloads). **Unless you know what you're doing, allow Anaconda to install in its prefered location, and to modify your path.**
- If you installed Anaconda in a terminal window, close that window now, as the changes to your path will only be active in the _next_ window you open.

#### 2. Install additional software

- In a **new** terminal window:

```
conda install -y basemap netCDF4 pandas=0.18.0 xarray
pip install plotly
```

You will also need a working `unzip` program. Please install it via your package manager or other means if you do not already have it.

#### 3. Download the tutorial files

- Visit the [GitHub page](https://github.com/p-madden/sea-2016-tutorial) for this tutorial, and either `git clone` the repo or download and expand the ZIP file.

#### 4. Connect to Jupyter

- In your terminal window, `cd` into the directory containing the tutorial files and run

  `jupyter notebook --notebook-dir=.`


- Your browser should open to Jupyter, and you should see a list of the `.ipynb` notebooks for the tutorial.

#### Updating the tutorial files

If you downloaded a ZIP file in step 3, discard the directory created when you extracted it, then download and extract a fresh ZIP file.

If you used `git clone` in step 3, cd into the directory created by that command and do `git checkout . && git pull`. Note that this will reset any changes you have made to the notebook files. Alternatively, you can rename (or delete) your old directory and do a new `git clone`.

_ **Note:** We will have removable media with the software-intstaller and tutorial files available at the tutorial for those who have problems installing the software ahead of time, but **please** try to prepare your software in advance!_
